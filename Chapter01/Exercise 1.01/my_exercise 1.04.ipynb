1.5 Generating a random list


```python
# import the random library
import random
```

```python
# Use randint
list_1 = [random.randint(0, 30) for x in range(0, 100)]
```

```python
list_1
```

```python
# Find the square root of the list
list_2 = [x**2 for x in list_1]
list_2
```

```python
# Import math
import math
```

```python
list_2 = [math.log(x + 1, 10) for x in list_2]
list_2
```
